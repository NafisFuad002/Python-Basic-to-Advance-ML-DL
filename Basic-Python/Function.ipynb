```python
# Defining a function
def fun():
  print("I'm a function")

fun()
fun()
```

```python
# Parameter & Argument
def fun(a, b): # a, b: called parameter
  print(a+b)

fun(10, 20) # 10, 20: called argument
```

```python
# return value
def fun(a,b):
  return a+b

res=fun(5,10)
print(res)
```

```python
# Default parameter
def fun2(name="User"):
    print("Hello", name)

fun2()
fun2("Nafis")
```

```python
# Keyword Arguments
def student(name, age):
    print(name, age)

student(age=20, name="Rahim")
```

```python
# Variable-length arguments
# *args (multiple values)
def fun3(*numbers):
    print(sum(numbers))

fun3(1,2,3,4)
```

```python
# Variable-length arguments
# **kwargs (key-value)
def info(**data):
    print(data)

info(name="Nafis", age=20)
```

```python
# Lambda Function
square = lambda x: x * x
print(square(5))
```

```python
# Generator
def my_generator():
  yield 1
  yield 2
  yield 3

for value in my_generator():
  print(value)
```

```python
# Practise Problems
# Factorial
number=int(input("Enter a number: "))
def fact(n):
  if n==0:
    return 1
  else:
    return n*fact(n-1)

res=fact(number)
print(res)
```

```python
# Recursion 2
# find sum of [1,3,5,7,9,11,13]
A=[1,3,5,7,9,11,13]
def add(a,k):
  if k<0:
    return 0
  else:
    return a[k]+add(a, k-1)

res=add(A,6)
print(res)
```
