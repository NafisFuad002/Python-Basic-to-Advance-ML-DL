```python
# for loop
for i in range(5):
    print(i,end=" ")
```

```python
# while loop
i = 0
while i < 5:
    print(i,end=" ")
    i += 1
```

```python
# nested loop
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# break
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")
```

```python
# continue
for i in range(7):
    if i == 2:
        continue
    print(i, end=" ")
```

```python
# pass
for i in range(7):
    pass
```
