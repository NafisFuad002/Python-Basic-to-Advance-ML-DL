```python
# if condition
a=10
if a==10:
  print("Yes, a is 10")
```

```python
# if-else
a=20
if a==10:
  print("Yes, a is 10")
else:
  print("No, a isn't 10")
```

```python
# if-elif-else
a=20
if a<10:
  print("a is less than 10")
elif a>10:
  print("a is greater than 10")
else:
  print("a is equal to 10")
```

```python
# nested if
A=100
if A>50:
  if A%2==0:
    print("A is greater than 50 and even")
  else:
    print("A is greater than 50 and odd")
else:
  if A%2==0:
    print("A is less than or equal to 50 and even")
  else:
    print("A is less than or equal to 50 and odd")
```

```python
# Ternary Operator
x=10
result = "Even" if x % 2 == 0 else "Odd"
print(result)
```

```python
# Logical Condition
x=5
if x > 0 and x < 10:
    print("Between 1-9")

y=10
if y<0 or y>=10:
    print("Yes")
```

```python
# Membership Condition
if 3 in [1,2,3]:
    print("Found")

if 5 not in [1,2,3]:
    print("Not Found")
```

```python
# match-case
day=2
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case _:
        print("Other day")
```
