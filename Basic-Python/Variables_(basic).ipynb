```python
# Some basics about variables
a=10
b=3
c="Hello World!"
print(a)
print(b)
print(c)
```

```python
# Casting
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print(x)
print(y)
print(z)
```

```python
# Legal variable names:
myvar = "John"
my_var = "John"
_my_var = "John"
myVar = "John"
MYVAR = "John"
myvar2 = "John"
print(myvar)
print(my_var)
print(_my_var)
print(myVar)
print(MYVAR)
print(myvar2)
```

```python
# Assign multiple values in multiple variables
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
```

```python
# Assign one value in multiple variables
x = y = z = "Orange"
print(x)
print(y)
print(z)
```

```python
# Operations
a=30
b=20
c=a+b
d=a-b
e=a*b
f=a/b # Quotient
g=a%b # Remainder
h=a//b # Quotient without floating-point number
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
```

```python
# Advance operation
print(25 ** (1/2)) # Square root
print(27 ** (1/3)) # Cube root

# print(100 ** (1/n)) # nth root

print(2 ** 3) # 2^3
```
